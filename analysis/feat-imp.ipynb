```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_train_imps = pd.read_csv('../gcn/logs/LPL-LCK-LCS-2020/train/feat_imp.csv')
```

```python
importances = raw_train_imps.T[0]
std = raw_train_imps.T[1]
indices = np.argsort(importances)[::-1]
fig, ax = plt.subplots()
plt.title("Feature importances: LPL-LPL (train) 2020")
plt.bar(range(10), 
        importances[indices][:10],
        color="r", 
        yerr=std[indices][:10], 
        align="center")
plt.xticks(range(10), raw_train_imps[[c for c in raw_train_imps.columns if c != 'label']].columns[indices[:10]])
plt.xticks(rotation=60)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
raw_val_imps = pd.read_csv('../gcn/logs/LPL-LCK-LCS-2020/val/feat_imp.csv')
```

```python
importances = raw_val_imps.T[0]
std = raw_val_imps.T[1]
indices = np.argsort(importances)[::-1]
fig, ax = plt.subplots()
plt.title("Feature importances: LPL-LCK (val) 2020")
plt.bar(range(10), 
        importances[indices][:10],
        color="r", 
        yerr=std[indices][:10], 
        align="center")
plt.xticks(range(10), raw_val_imps[[c for c in raw_val_imps.columns if c != 'label']].columns[indices[:10]])
plt.xticks(rotation=45)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
